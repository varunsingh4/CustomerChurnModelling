```python
# Droping useless columns
df.drop(columns=['RowNumber','CustomerId','Surname'],axis=1,inplace=True)
df.head()
```

```python
df.Geography.value_counts()
```

```python
#Coverting to categorical features
df['Geography']=df['Geography'].map({'France':0,'Germany':1,'Spain':2})
df['Gender']=df['Gender'].map({'Female':0,'Male':1})
```

```python
df.head()
```

```python
#Creating features and labels
from tensorflow.keras.utils import to_categorical
X=df.drop('Exited',axis=1)
Y=to_categorical(df.Exited)
```

```python
#Splitting into train-test split
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.25)
```

```python
#Scaling data
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.fit_transform(X_test)
```

```python
import keras
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import Dropout
from keras.layers import BatchNormalization
```

```python
#Initializing Ann
model=Sequential()
#First Input Layer and Hidden layer with 10 neurons
model.add(Dense(10,kernel_initializer='normal',activation='relu',input_shape=(10,)))
#Add Dropout
model.add(Dropout(rate=0.1))
model.add(BatchNormalization())
#Adding 3rd and Hidden layer
model.add(Dense(7,kernel_initializer='normal',activation='relu'))
model.add(Dropout(rate=0.1))
model.add(BatchNormalization())
#Adding output layer
model.add(Dense(2,kernel_initializer='normal',activation='sigmoid'))
```

```python
model.summary()
```

```python
#Compiling model
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
#Fitting model to the training set
model_history=model.fit(X_train,Y_train,validation_split=0.2,validation_data=(X_test,Y_test),epochs=100)
```

# Visualizing training data and validation loss

```python
plt.figure(figsize=(12,6))
train_loss=model_history.history['loss']
val_loss=model_history.history['val_loss']
epoch=range(1,101)
sns.lineplot(epoch,train_loss,label='Training Loss')
sns.lineplot(epoch,val_loss,label='Validation Loss')
plt.title('Training and Validation Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()
```

# Visualizing training and validation accuracy

```python
plt.figure(figsize=(15,6))
train_loss=model_history.history['accuracy']
test_loss=model_history.history['val_accuracy']
epoch=range(1,101)
sns.lineplot(epoch,train_loss,label='Training accuracy')
sns.lineplot(epoch,val_loss,label='Validation accuracy')
plt.title('Training and Validation Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.show()
```

```python
acc = model.evaluate(X_test, Y_test)[1]

print(f'Accuracy of model is {acc}')
```
